Exercise 1. k Selection via Elbow Method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
```

```python
df = pd.read_csv("/content/imdb_movies_2000to2022.actorXgenre.csv",
                 index_col="actor_id").apply(pd.to_numeric, errors='coerce').fillna(0)

df=df.reindex(sorted(df.columns), axis=1)
```

```python
inertia_scores=[]
for test_k in range(2,31):
  print("test k:",test_k,)
  tmp_model=MiniBatchKMeans(
      n_clusters=test_k,
      n_init=16,
      max_iter=2048,
      tol=0.5,
      reassignment_ratio=0.5,
      random_state=42
  )
  tmp_model.fit(df)
  scores=tmp_model.inertia_
  inertia_scores.append((test_k,scores))
```

```python
#Create intertia dataframe
inertia_df=pd.DataFrame(inertia_scores, columns=["k","inertia"])

#Display graph of k clusters by inertia
fig=plt.figure(figsize=(12,7))
ax=fig.add_subplot(1,1,1)

inertia_df.sort_values(by="k").plot("k","inertia",marker="o",ax=ax)

ax.set_xlabel("k (number of clusters)")
ax.set_ylabel("inertia")
ax.set_title("Elbow Method")
ax.grid(True, alpha=0.3)
plt.show()
```
